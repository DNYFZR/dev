```python
import numpy as np, matplotlib.pyplot as plt
from seaborn import set_theme
set_theme()
```

```python
# === Probability ===
first = []
second = []

for _ in range(50000):
    # Randomise order
    x = np.array([0,1,0])
    np.random.shuffle(x)

    # Set ids on random order
    idx = [i for i in range(len(x))]

    # Make first choice
    choice = np.random.choice(idx)
    first.append(x[choice])

    # Identify remaining
    remaining = np.delete(idx, choice, 0)

    # Remove a zero option
    if np.sum(x[remaining]) == 0:
        remove = np.random.choice(remaining)
    else:
        for val in remaining:
            if x[val] == 0:
                remove = val

    # Review switch
    switch = np.delete(remaining, np.where(remove > 0, 1, 0), 0)[0]
    second.append(x[switch])

print((np.mean(first), np.mean(second)))

# Plot
plt.figure(figsize=(10, 3))

plt.plot([np.mean(first[:n]) for n in range(len(first)) if n > 0])
plt.plot([np.mean(second[:n]) for n in range(len(second)) if n > 0])
```
